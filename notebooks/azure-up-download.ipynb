```python
import os
import dotenv
import pandas as pd
from azure.storage.blob import BlobServiceClient


dotenv.load_dotenv("../src/.env")
```

```python
input_pdf_path = "../resources/potential_origin_pdf_directory_2023"
input_pdf_file_list = os.listdir(input_pdf_path)
```

```python
res = pd.read_excel("./potential_reports_cleaned_240605.xlsx")
res.head()
```

```python
# Your Azure storage connection string# connection string = "Your storage connection string!
connect_str = os.getenv('AZURE_STORAGE_CONNECTION_STRING')
container_name = "nsark"

# # Create the Blobserviceclient obiectblob service 
blob_service_client = BlobServiceClient.from_connection_string(connect_str)
container_client = blob_service_client.get_container_client(container_name)
```

```python
# 上传Blob文件
for i in range(len(res)):
    if res["link_filename"][i] in input_pdf_file_list:
        link_file_name = res["link_filename"][i]
        file_path = f"../resources/potential_origin_pdf_directory_2023/{link_file_name}"
        
        code = res["code"][i]
        document_title = res["document_title"][i]
        fiscal_year = res["fiscal_year"][i]
        
        blob_name = f"audit_files/{code}/{code}_FY{fiscal_year}_{document_title}.pdf"
        
        try:
            with open(file_path, "rb") as data:
                blob_client = container_client.upload_blob(blob_name, data, overwrite=False)
            print(f"{blob_name} Blob uploaded successfully.")
        except FileNotFoundError:
            print("The file to upload was not found.")
        except Exception as e:
            print(f"Error uploading blob: {blob_name}", e)
```

```python
print("\nListing blobs...")

blob_list = container_client.list_blobs(name_starts_with="audit_files/0939.HK/")

for blob in blob_list:
    print(blob.name)
```

```python
# 下载Blob文件
report_pdf = pd.read_excel("../resources/new_reports_240607.xlsx")
report_pdf.head()
```

```python
save_pdf_path = "../resources/eric_pdf"
save_pdf_file_list = os.listdir(save_pdf_path)
```

```python
for i in range(len(report_pdf)):
    blob_name = report_pdf["path"][i]
    file_name = blob_name.split("/")[-1]
    try:
        with open(os.path.join(save_pdf_path, file_name), "wb") as download_file:
            blob_client = container_client.get_blob_client(blob_name)
            download_stream = blob_client.download_blob()
            download_stream.readinto(download_file)
            print(f"{blob_name} Blob download successfully.")
    except Exception as e:
        print(f"Error download blob: {blob_name}", e)
```
