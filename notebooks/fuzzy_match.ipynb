```python
import os
import json
import openai
import dotenv
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
data_df = pd.read_csv("../resources/potential_assessment_2023_info@20240604.csv")
```

```python
data_df.head()
```

```python
data_df["company_match_score"] = None

for i in range(len(data_df)):
    if pd.isna(data_df["company"][i]) == False:
        res = fuzz.ratio(str(data_df["name"][i]), str(data_df["company"][i]))
        # print(res)
        data_df.loc[i, "company_match_score"] = res
    
```

```python
data_df.to_csv("./potential_assessment_2023_info@20240605.csv", index_label=None)
```
